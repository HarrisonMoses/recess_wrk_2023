```python
#We can perform mathematicalm operations like addition, multiplication, deveision among others
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])


addition = series1 + series2
subtraction = series1 - series2
multiplication = series1 * series2
division = series1 / series2

print("Addition:", )
print(addition)
print("Subtraction:")
print(subtraction)
print("Multiplication:")
print(multiplication)
print("Division:")
print(division)
```

```python
#if i want to sum up the most appearance of the occurance

series =  Africa['HomeTeamGoals']
most_frequent = series.mode()[0]
series = series.apply(lambda x: x if x == most_frequent else 'Other')
count_most_frequent = series.value_counts()[most_frequent]

print("Series with replaced values:")
print(series)
print("Count of the most frequent value:", count_most_frequent)
```
