```python
# Drop the row that has school as a null
df = df.dropna(subset=['SCHOOL'])
```

```python
# Replace NaN with zero
df.fillna(0)
df.head()
```

```python
# Confirm that there are nolonger any null values.
df.info()
```

```python
candidates_2011 = df[df['YEAR'] == 2011]['TOTAL CANDIDATES'].sum()
print("Candidates in 2011:", candidates_2011)
candidates_2012 = df[df['YEAR'] == 2012]['TOTAL CANDIDATES'].sum()
print("Candidates in 2012:", candidates_2012)
candidates_2013 = df[df['YEAR'] == 2013]['TOTAL CANDIDATES'].sum()
print("Candidates in 2013:", candidates_2013)
candidates_2014 = df[df['YEAR'] == 2014]['TOTAL CANDIDATES'].sum()
print("Candidates in 2014:", candidates_2014)
candidates_2015 = df[df['YEAR'] == 2015]['TOTAL CANDIDATES'].sum()
print("Candidates in 2015:", candidates_2015)
candidates_2016 = df[df['YEAR'] == 2016]['TOTAL CANDIDATES'].sum()
print("Candidates in 2016:", candidates_2016)
```

```python
# Plot a line graph
years = [2011, 2012, 2013, 2014, 2015, 2016]
candidates = [candidates_2011, candidates_2012, candidates_2013, candidates_2014, candidates_2015, candidates_2016]

plt.plot(years, candidates)

# Add a title and labels to the axes
plt.title("Line graph showing changes in the total number of candidates who sat for UCE.")
plt.xlabel("Year")
plt.ylabel("Number of Candidates")
plt.show()
```

```python
div1_2011 = df[df['YEAR'] == 2011]['TOTAL DIV 1'].sum()
print("Division ones in 2011:", div1_2011)
div1_2012 = df[df['YEAR'] == 2012]['TOTAL DIV 1'].sum()
print("Division ones in 2012:", div1_2012)
div1_2013 = df[df['YEAR'] == 2013]['TOTAL DIV 1'].sum()
print("Division ones in 2013:", div1_2013)
div1_2014 = df[df['YEAR'] == 2014]['TOTAL DIV 1'].sum()
print("Division ones in 2014:", div1_2014)
div1_2015 = df[df['YEAR'] == 2015]['TOTAL DIV 1'].sum()
print("Division ones in 2015:", div1_2015)
div1_2016 = df[df['YEAR'] == 2016]['TOTAL DIV 1'].sum()
print("Division ones in 2016:", div1_2016)
```

```python
# Plot a line graph
years = [2011, 2012, 2013, 2014, 2015, 2016]
div1 = [div1_2011, div1_2012, div1_2013, div1_2014, div1_2015, div1_2016]

plt.plot(years, div1)

# Add a title and labels to the axes
plt.title("Line graph showing changes in the number of division ones got by students in the respective years.")
plt.xlabel("Year")
plt.ylabel("Division ones")
plt.show()
```

**SECOND DATASET.**

```python
# Loading the second dataset
df2 = pd.read_csv('MPsProfileInformation10thParliament.csv', encoding='latin-1')
df2
```

```python
# Drop column No.
df2 = df2.drop('No.', axis=1)
df2.head(2)
```

```python
df2.columns
```

```python
df2.info()
```

```python
# Drop columns Committee_4 and Role on that committee since only 2 of 439 rows have data. The rest have nulls.
df2 = df2.drop('Committee_4', axis=1)
df2 = df2.drop('Role on Committee 4', axis=1)
df2.head(2)
```

```python
df2.head(10)
```

```python
# Handling wrong format for date in the row at position 9   i.e   07/071983
df2['Date of Birth'][9] = '07/07/1983'
```

```python
# Invalid date  09/0/1971  in the row at position 93
df2['Date of Birth'] = df2['Date of Birth'].replace('09/0/1971', np.nan)
```

```python
# Converting all the dates to the correct date format.
df2['Date of Birth'] = pd.to_datetime(df2['Date of Birth'], format='mixed')
df2.head(10)
```

```python
df2['Political Party'].unique()
```

```python
# In the political party column, there are two different values both referring to the same party: 
# Replace 'INDEPENDENT' with 'Independent'
df2['Political Party'] = df2['Political Party'].replace('INDEPENDENT', 'Independent')
```

```python
# Filling null values in the Political Party column with 'NONE'
df2['Political Party'] = df2['Political Party'].fillna('NONE')
```

```python
# Count the frequency of each political party
party_counts = df2['Political Party'].value_counts()

# Create the pie chart
plt.pie(party_counts, labels=party_counts.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie chart showing the distribution of political parties of MPs.')
plt.show()
```

```python
df2['Marital status'].unique()
```

```python
# Define the allowed values
allowed_values = ['Married', 'Single', 'Widowed', 'Divorced']

# Replace values other than the allowed ones with 'Private'
df2['Marital status'] = df2['Marital status'].where(df2['Marital status'].isin(allowed_values), 'Private')
```
